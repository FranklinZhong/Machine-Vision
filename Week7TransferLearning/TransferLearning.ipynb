Using Pre-trained model for transfer learning

Import the libraries

```python
from keras.applications import VGG16, InceptionV3, ResNet50
from keras.layers import Flatten, Dense
from keras.models import Model
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

Dataset preprocessing

```python
# Pre-trained model train 1k classes in Imagenet. 
# But we only have 3 classes. The classes are different.
# So we should ignore the FC. Because it's not the same.
# And we have to create our own FC.
# The output should look like Activation map.
# The output should avalible for the FC we create.

# base_model = VGG16(include_top = False)
# first_model = InceptionV3()

base_model = ResNet50(include_top = False, 
                      weights = "imagenet", 
                      input_shape=(128, 128, 3))
x = Flatten()(base_model.output)
x = Dense(100, activation='relu')(x)
output_layer = Dense(3, activation='softmax')(x)

# Don't change any weights in base_model. Any change will destroy the model performance
# Freeze the base model not update the weights during training
for layer in base_model.layers:
    layer.trainable = False

model = Model(inputs = base_model.input, outputs = output_layer)

model.summary()
```
